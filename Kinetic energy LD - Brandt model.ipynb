```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def kinetic_energy_Brandt_interactive(h_max):
    
    # Generate data from 0 to the user-defined h_max
    h = np.linspace(0, h_max, 200)
    
    # Brandt Equation: KE = 15.8 * h^(1/2) - 5.87
    ke = 15.8 * np.sqrt(h) - 5.87
    ke[h>8] =  15.8 * np.sqrt(8) - 5.87
    ke[h<0.15] = 0
    
    plt.figure(figsize=(9, 5))
    plt.plot(h, ke, linewidth=2.5, color='forestgreen', label='Brandt (1989)')
    
    plt.title('Kinetic Energy of Leaf Drainage (Brandt Equation)', fontsize=14)
    plt.xlabel('Height $h$ (m)', fontsize=12)
    plt.ylabel('Kinetic Energy $KE_{LD}$ ($J \cdot m^{-2} \cdot mm^{-1}$)', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

# Create the interactive slider
interact(kinetic_energy_Brandt_interactive, 
         h_max=FloatSlider(value=15.0, min=0.1, max=30.0, step=0.5, description='h_max:'))
```
