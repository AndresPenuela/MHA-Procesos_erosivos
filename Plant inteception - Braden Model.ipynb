```python
import numpy as np
import matplotlib.pyplot as plt

def plant_interception_Braden(a_cm, LAI, rain_min_mm, rain_max_mm):
    """
    Computes and plots plant interception based on the Braden (1985) model.
    
    Formula: PI = a * LAI * (1 - 1 / (1 + rain / (3 * a)))
    """
    # Create an array of rainfall values in mm
    rain_mm = np.linspace(rain_min_mm, rain_max_mm, 100)
    
    # Convert rainfall from mm to cm for the formula
    rain_cm = rain_mm / 10.0
    
    # Braden (1985) formula (result is in cm/day)
    PI_cm = a_cm * LAI * (1 - (1 / (1 + (rain_cm / (3 * a_cm)))))
    
    # Convert PI from cm back to mm for plotting
    PI_mm = PI_cm * 10.0
    
    # Plotting
    plt.figure(figsize=(8, 5))
    plt.plot(rain_mm, PI_mm, label=f'a={a_cm}, LAI={LAI}', color='forestgreen', lw=2)
    plt.title('Plant Interception (Braden Model)', fontsize=14)
    plt.xlabel('Daily Rainfall (mm)', fontsize=12)
    plt.ylabel('Plant Interception (mm)', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()

# Execute the function with the specified parameters
plant_interception_Braden(a_cm=0.5, LAI=3, rain_min_mm=10, rain_max_mm=60)
```
